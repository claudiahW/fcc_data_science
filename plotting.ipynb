```python
# reading the csv file 
df_population_raw = pd.read_csv(r'C:\Users\user\Downloads\population_total.csv')
```

```python
df_population_raw 

#dropping null values 

df_population_raw.dropna(inplace=True)
```

```python
df_pivot = df_population_raw.pivot_table(index='year', columns='country', values='population')
df_pivot
```

# 1. Lineplot 

```python
df_pivot.plot(kind='line',ylabel='Population',xlabel='Year', title='Population (1955-2020)',
              figsize=(8,4))
```
