# 1. Lambda function

```python
# basic function 
def sum_values(a,b):
    x = a+b
    return x 
```

```python
sum_values(2,-3)
```

```python
# using lambda function (one liner)
sum_values_lambda = lambda a, b: a + b

sum_values_lambda(2,3)
```

```python
# use lambda function to convert 'Name' series to uppercase 
df_titanic['Name'].apply(lambda x:x.upper())
```
