# making static and interactive pivot tables 

```python
# reading new csv file 

df_population_raw = pd.read_csv(r'C:\Users\user\Downloads\population_total.csv')
```

```python
df_population_raw
```

# making a pivot table 

```python
# dropping null values 
df_population_raw.dropna(inplace=True)
```

```python
# making a pivot table 
df_pivot = df_population_raw.pivot(index='year',
                        columns='country',
                        values='population')
```

```python
df_pivot = df_pivot[['United States', 'Iran', 'Kenya','India', 'China']]
```

```python
# showing the pivot table
df_pivot
```

# 2 Lineplot

```python
df_pivot.plot(kind='line', xlabel='Year', ylabel='Population', title='Population(1955-2020)')
```

# 3. Barplot 

## 3.1 Single Barplot 

```python
df_pivot.index
```

```python
# selecting only one year 

df_pivot_2020 =df_pivot[df_pivot.index.isin([2020])]
```

```python
# show new dataframe 
df_pivot_2020
```

```python
# transpose dataframe from rows to columns and vice versa
df_pivot_2020 = df_pivot_2020.T
```

```python
# making barplot 
df_pivot_2020.plot(kind='bar', color='orange',
                   xlabel='Year', ylabel='Population',
                   title='Population 2020')
```

## 3.2 Barplot grouped by "n" variables

```python
df_pivot
```

```python
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
df_pivot_sample.plot(kind='bar', xlabel='Year', ylabel='Population', title='Population growth over time')
```
