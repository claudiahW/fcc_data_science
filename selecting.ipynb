# 1 Creating a Conditional column from more than 2 choices 

```python
# import necessary libraries and packages 
import numpy as np, scipy as sp, pandas as pd
```

```python
# load the file as csv 
df_crops = pd.read_csv(r"C:\Users\user\Downloads\Kenya_Crops_Dataset.csv")
```

```python
df_crops.columns
```

```python
df_crops['Profit (KES)'].min()
```

```python
# create an array based on profits made

#create " conditions" and "values" variables

conditions = [
    df_crops['Profit (KES)']>10000000,
    (df_crops['Profit (KES)']>5000000) & (df_crops['Profit (KES)']<10000000),
    (df_crops['Profit (KES)']>0) & (df_crops['Profit (KES)']<500000),
    (df_crops['Profit (KES)']<= 0) 
]

values =['Great', 'Okay', 'Average', 'Poor']
```

```python
# set it to a new column 

df_crops['Profit_tiers']=np.select(conditions, values)
```

# 2. isin(): Multiple filtering


```python
# find coffee or tea 

filter1 = df_crops['Crop Type'].isin(['Tea', 'Coffee'])
filter2 = df_crops['County'].isin(['Nairobi', 'Malindi'])
```

```python
# link filter1 and filter2 using an operator 

df_crops[filter1 & filter2]
```
