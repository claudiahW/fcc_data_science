# 1. Create a column based on 1 condition:
np.where()

```python
# create an array based on Revenue tiers

np.where(df_crops['Revenue (KES)'] > 4500000, 'Affluent', 'Basic')

df_crops['Tax_tier'] = np.where(df_crops['Revenue (KES)'] > 4500000, 'Affluent', 'Basic')
```

```python
#list available columns 
df_crops.columns
```

```python
# check the total number of columns in the newly added Tax Tier
df_crops.value_counts('Tax_tier')
```

```python
# list either Nakuru or Eldoret counties 
df_crops['County']=='Nakuru'
df_crops['County']=='Eldoret'

(df_crops['County']=='Nakuru' ) | (df_crops['County']=='Eldoret')

df_crops[(df_crops['County']=='Nakuru' ) | (df_crops['County']=='Eldoret')]
```
